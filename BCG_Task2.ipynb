## These are all the frequently or possible Question that can be asked and answered as well.

## These Question are written using if-else statments, these will be compared with the user query and the bot will responds accordingly.

```python
def respond_to_query(company, year):
    query = input("Enter the Question:").lower()

    if query.lower() == "what is the total revenue?":
        Total_ravenue = df[(df['Year']==year) & (df["Company"]==company)]['Total ravenue (million$)'].values[0]
        return f"The total revenue of {company} in {year} is {Total_ravenue} millions dollar"

    elif query.lower() == "what is the total income?":
        Total_income = df[(df['Year']==year) & (df["Company"]==company)]['Net Income (million$)'].values[0]
        return f"The total income of {company} in {year} is {Total_income} millions dollar"

    elif query.lower() == "what is the total assets?":
        assets = df[(df['Year']==year) & (df["Company"]==company)]['Total Assets (million$)'].values[0]
        return f"The total assets of {company} in {year} is {assets} millions dollar"

    elif query.lower() == "what is the total liability?":
        liabilities = df[(df['Year']==year) & (df["Company"]==company)]['Total Liabilities (million$)'].values[0]
        return f"The total liability of {company} in {year} is {liabilities} millions dollar"


    elif query.lower() == "what is the total cash flow?":
        cash_flow = df[(df['Year']==year) & (df["Company"]==company)]['Cash Flow (million$)'].values[0]
        return f"The total cash flow by operating activities of {company} in {year} is {cash_flow} millions dollar"

    elif query.lower() == "what is the revenue growth?":
        revenue_growth = df[(df['Year']==year) & (df["Company"]==company)]['Revenue Growth (%)'].values[0].round(2)
        return f"The revenue growth of {company} in {year} is {revenue_growth}%"

    elif query.lower() == "what is the net income growth?":
        income_growth = df[(df['Year']==year) & (df["Company"]==company)]['Net Income Growth (%)'].values[0].round(2)
        return f"The net income growth of {company} in {year} is {income_growth}%"

    elif query.lower() == "what is the assets growth?":
        assets_growth = df[(df['Year']==year) & (df["Company"]==company)]['Assets Growth (%)'].values[0].round(2)
        return f"The assets growth of {company} in {year} is {assets_growth}%"

    elif query.lower() == "what is the liabilities growth?":
        liabilities_growth = df[(df['Year']==year) & (df["Company"]==company)]['Liability Growth (%)'].values[0].round(2)
        return f"The liabilities growth of {company} in {year} is {liabilities_growth}%"

    elif query.lower() == "what is the cash flow growth?":
        cash_flow_growth = df[(df['Year']==year) & (df["Company"]==company)]['Cash Flow Growth (%)'].values[0].round(2)
        return f"The cash flow growth of {company} in {year} is {cash_flow_growth}%"

    else:
        return "I'm sorry, I didn't understand that question, can you ask something else."
```

## This is the main function which will calls the function of answers(FAQs). This bot takes name of the company and year as input and gives the according answer of that query. if user enter bye, exit, quit, the bot will end.

```python
def main():
    while True:
        try:

            user_company = input("\nEnter the Company's Name: ").capitalize()
            exit = ["exit", "quit", "goodbye", "bye"]

            if user_company.lower() in exit:
                print("Chatbot: Goodbye! Have a great day!")
                break

            user_year = input("Enter the year:")
            if user_year in exit:
                print("Chatbot: Goodbye! Have a great day!")
                break
            user_year = int(user_year)
            companies = ["Apple", "Microsoft", "Tesla"]
            years = [2021,2022,2023]
            if user_company in companies and user_year in years:

                response = respond_to_query(user_company, user_year)
            #response = financial_chatbot(user_company, int(user_year))
                print("Chatbot:", response)
            else:
                print("please enter the correct name or year")
                main()

        except KeyboardInterrupt:
            print("\nChatbot: Bye! See you next time. (interrupted)")
            break

if __name__ == "__main__":
    print("To finish the chat, In either box type: exit or quit or goodbye or bye.\n")
    print("Welcome to financial chatbot!\n")
    print("I can only assist you with big companies like Apple, Microsoft and Tesla of year 2021, 2022, and 2023")

    print("How may i help you, please try to avoid spelling mistake.")
    main()
```
